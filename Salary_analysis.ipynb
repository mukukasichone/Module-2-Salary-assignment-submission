```python
import pandas as pd
import csv
import os 
import zipfile
```

```python
data = pd.read_csv("Salaries.csv")
data.head()
```

```python
employee_dict = data.groupby("EmployeeName").first().to_dict(orient="index")
```

```python
def get_employee_details(name):
    try:
        name = name.upper()
        employee = employee_dict[name]
        
        return {
            "Name": name,
            "JobTitle": employee["JobTitle"],
            "BasePay": employee["BasePay"],
            "OvertimePay": employee["OvertimePay"],
            "TotalPay": employee["TotalPay"]
        }
    
    except KeyError:
        return "Employee not found."
```

```python
print(get_employee_details("GARY JIMENEZ"))
print(get_employee_details("unknown"))
```

```python
employee = get_employee_details("GARY JIMENEZ")

if isinstance(employee, dict):
    with open("employee_details.csv", "w", newline="") as file:
        writer = csv.writer(file)
        
        writer.writerow(["Name", "JobTitle", "BasePay", "OvertimePay", "TotalPay"])
        
        writer.writerow([
            employee["Name"],
            employee["JobTitle"],
            employee["BasePay"],
            employee["OvertimePay"],
            employee["TotalPay"]
        ])
```

```python
folder = "Employee Profile"

if not os.path.exists(folder):
    os.makedirs(folder)

os.replace("employee_details.csv", f"{folder}/employee_details.csv")
```

```python
with zipfile.ZipFile("employee_profile.zip", "w") as zipf:
    zipf.write("Employee Profile/employee_details.csv", arcname="employee_details.csv")
```
